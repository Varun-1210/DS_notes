````python
# importing libraries

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd 
````

````python
# to see or print the plot in jupyter notebook, one has to use the below line

%matplotlib inline 
````

````python
# variables

x = np.linspace(0,5,11)
y = x ** 2
print(x,y,sep='\n')
````

````python
# basic example

plt.plot(x,y,'r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title') 
````

````python
# Multiple plot

plt.subplot(1,2,1) # row, column, reference
plt.plot(x,y,'r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('First title')

plt.subplot(1,2,2) # row, column, reference
plt.plot(y,x,'g')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Second title')
````

## Object oriented process

````python
# to initiate blank canvas 
fig = plt.figure() 

# to add the axes to the canvas
axes = fig.add_axes([0,0,1,1]) # [x start point, y start point, % of canvas to use as hight, % of canvas to use as width]
axes.plot(x,y)

# to add labels
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
````

````python
# blank canvas
fig = plt.figure()

# adding two axies
axes_1 = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.1,0.5,0.4,0.4])

# plotting for first axis
axes_1.plot(x,y,'r')

# plotting for second axis
axes_2.plot(y, x,'g')

# labels for both the axis
axes_1.set_xlabel('X label')
axes_1.set_ylabel('Y label')
axes_2.set_xlabel('X label')
axes_2.set_ylabel('Y label')
````

## Object oriented process - Subplots

````python
# Creating subplots
# this process will create axis and fig automatically
fig,axes = plt.subplots() 

# adding plots to axis
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
````

````python
# to create two plots
fig,axes=plt.subplots(nrows=1,ncols=2) # to create two plots in one row and two columns

# adding plot
axes[0].plot(x,y)
axes[1].plot(y,x)

# adding labels
axes[0].set_xlabel('X label')
axes[0].set_ylabel('Y label')

axes[1].set_xlabel('X label')
axes[1].set_ylabel('Y label')

# adding title
axes[0].set_title('1 Table')
axes[1].set_title('2 Table')
````
